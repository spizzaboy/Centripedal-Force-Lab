```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt


Slope = np.array([9.596,9.952,10.10,9.928])
Period_Squared = np.array([2.1112,1.5625,1.3133,0.77440])
Error_Period = np.array([0.081, 0.065, 0.069, 0.035])
dy = Error_Period
x = Slope
y = Period_Squared

def Delta(x, dy):
    D = (sum(1/dy**2))*(sum(x**2/dy**2))-(sum(x/dy**2))**2
    return D
 
D = Delta(x, dy)

dm = np.sqrt(1/D*sum(1/dy**2)) #error in slope
db = np.sqrt(1/D*sum(x**2/dy**2)) #error in intercept

plt.figure(figsize=(15,5))
 
plt.plot(Slope, Period_Squared, color='green', linestyle='-')
plt.scatter(Slope, Period_Squared, color='black', marker='o')
plt.xlabel('Newton Meters (N*m)')
plt.ylabel('Period_Squared (s^2) ')
plt.title('Centripedal Force Plot - Torque vs Period Squared')
plt.annotate('Error in Slope (N*m) = {value:.{digits}E}'.format(value=dm, digits=1),
             (0.05, 0.85), xycoords='axes fraction')
plt.show()

```
